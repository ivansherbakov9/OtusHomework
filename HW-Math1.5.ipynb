```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.random.rand(100, 100)
```

```python
U, S, Vt = np.linalg.svd(A)
```

```python
def error(A, Ar):
    return np.linalg.norm(A - Ar)**2
```

```python
errors = []

for r in range(2, 51):
    Ar = np.dot(U[:, :r], np.dot(np.diag(S[:r]), Vt[:r, :]))
    err = error(A, Ar)
    errors.append(err)

plt.plot(range(2, 51), errors)
plt.xlabel("Ранг")
plt.ylabel("Ошибка");
```
